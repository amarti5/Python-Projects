```python
tickers = ['USDJPY=x', 'AUDUSD=x', 'BTCUSD=x', 'EURUSD=x', 'GE']

new_data = pd.DataFrame()
for t in tickers:
    new_data[t] = wb.DataReader(t, data_source='yahoo', start='1995-1-1')['Adj Close']
```
